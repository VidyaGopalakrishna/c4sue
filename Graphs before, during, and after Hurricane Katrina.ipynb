```python
#Important Dates
#Before Katrina = 2004.csv
#During Katrina = 2005.csv
#After Katrina = 2006.csv
#Hurricane Katrina = Aug 23 2005 - Aug 31 2005
#Gulf Coast States = Texas, Alabama, Mississippi, Louisiana, Florida
```

```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df04 = pd.read_excel('2004.xls')
df05 = pd.read_excel('2005.xls')
df06 = pd.read_excel('2006.xls')
```

```python

PersonalIncome04 = pd.DataFrame(df04.loc[df04['LineCode'] == 1],columns=['GeoName', 'data'])
Population04 = pd.DataFrame(df04.loc[df04['LineCode'] == 2],columns=['GeoName', 'data'])
PerCapita04 = pd.DataFrame(df04.loc[df04['LineCode'] == 3],columns=['GeoName', 'data'])
```

```python
PersonalIncome05 = pd.DataFrame(df05.loc[df05['LineCode'] == 1],columns=['GeoName', 'data'])
Population05 = pd.DataFrame(df05.loc[df05['LineCode'] == 2],columns=['GeoName', 'data'])
PerCapita05 = pd.DataFrame(df05.loc[df05['LineCode'] == 3],columns=['GeoName', 'data'])
```

```python
PersonalIncome06 = pd.DataFrame(df06.loc[df06['LineCode'] == 1],columns=['GeoName', 'data'])
Population06 = pd.DataFrame(df06.loc[df06['LineCode'] == 2],columns=['GeoName', 'data'])
PerCapita06 = pd.DataFrame(df06.loc[df06['LineCode'] == 3],columns=['GeoName', 'data'])
```

```python
PersonalIncome04.shape
PersonalIncome04 = PersonalIncome04[1::]
PersonalIncome05 = PersonalIncome05[1::]
PersonalIncome06 = PersonalIncome06[1::]

Population04 = Population04[1::]
Population05 = Population05[1::]
Population06 = Population06[1::]

PerCapita04 = PerCapita04[1::]
PerCapita05 = PerCapita05[1::]
PerCapita06 = PerCapita06[1::]
```

```python
#Personal Income 2004
fig,plot = plt.subplots(figsize=(20, 20))
width = 2.25
ind = np.arange(len(PersonalIncome04.data))
plot.bar(PersonalIncome04.GeoName,PersonalIncome04.data, color="blue")
#plot.set_yticks(ind+width/2)
#plot.set_yticklabels(PersonalIncome04.data, minor=False)
#plot.set_xticks(ind/2)
plot.set_xticklabels(PersonalIncome04.GeoName,minor=False,fontsize = 20)
plt.title("Annual Personal Income in each state in 2004")
plt.xlabel('States')
plt.ylabel('Millions of Dollars')
plt.savefig("PersonalIncome04.png")
for i in plot.patches:
    # get_x pulls left or right; get_height pushes up or down
    plot.text(i.get_x()+.04, i.get_height()+12000, \
            str(round((i.get_height()), 2)), fontsize=36, color='black')
```

```python
#Personal Income 2006
fig,plot = plt.subplots(figsize=(20, 20))
width = 2.25
ind = np.arange(len(PersonalIncome06.data))
plot.bar(PersonalIncome06.GeoName,PersonalIncome06.data, color="blue")
#plot.set_yticks(0,PersonalIncome06.max()-1)
#plot.set_yticklabels(PersonalIncome06.data, minor=False)
#plot.set_xticks(ind/2)
plot.set_xticklabels(PersonalIncome06.GeoName,minor=False,fontsize = 20)
plt.title("Annual Personal Income in each state in 2006")
plt.xlabel('States')
plt.ylabel('Millions of Dollars')
plt.savefig("PersonalIncome06.png")
# set individual bar lables using above list
for i in plot.patches:
    # get_x pulls left or right; get_height pushes up or down
    plot.text(i.get_x()+.04, i.get_height()+12000, \
            str(round((i.get_height()), 2)), fontsize=36, color='black')
```

```python
#Personal Income Comparisons
PI = pd.concat([PersonalIncome04,PersonalIncome05,PersonalIncome06], axis=0, ignore_index=False)
PI['Years'] = (len(PersonalIncome04.data)*(2004,) + len(PersonalIncome05.data)*(2005,) + len(PersonalIncome06.data)*(2006,))
PI
combined = sns.factorplot(x='GeoName', y='data', hue="Years",kind='bar', data=PI)
combined.set_axis_labels('States', 'Millions of Dollars')
plt.subplots_adjust(top=0.9)
combined.fig.suptitle('Total Personal Income over 3 years')
```
